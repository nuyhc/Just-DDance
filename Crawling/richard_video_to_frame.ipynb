```python
file_path = './[K-Fancam] 레드벨벳 아이린 직캠 Feel My Rhythm (Red Velvet IRENE Fancam) l @MusicBank 220325.mp4'
```

```python
# video to frame 수정 코드 

vidcap = cv2.VideoCapture(file_path)
success, image = vidcap.read()

count = 0

while success:
    cv2.imwrite("./frame/frame%d.jpg" % count, image)     # save frame as JPEG file"
    success, image = vidcap.read()
    print('Read a new frame: ', success)
    count += 1
```

```python
video = cv2.VideoCapture(file_path)

if not video.isOpened():
    print('Could not open video')
    exit()
```

```python
length = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(video.get(cv2.CAP_PROP_FPS))

print(length, width, height, fps)
```

```python
try:
    if not os.path.exists(file_path[:-4]):
        os.makedirs(file_path[:-4])
except OSError:
    print('Error: Creating directory of data')
    exit()
```

```python
count = 0

while video.isOpened():
    ret, frame = video.read()
    if(int(video.get(1))%fps == 0):
        cv2.imwrite(file_path[:-4] + '/' + str(count) + '.jpg', frame)
        count += 1
video.release()
```
